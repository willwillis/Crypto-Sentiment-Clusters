```python
import os
from datetime import datetime, timedelta
import time
```

```python
API_KEY = os.getenv("TWITTER_API_KEY")
API_SECRET_KEY = os.getenv("TWITTER_SECRET_KEY")
DEV_ENVIRONMENT_LABEL = 'fintech'
API_SCOPE = '30day'  # 'fullarchive' for full archive, '30day' for last 31 days

SEARCH_QUERY = 'bitcoin lang:en'
RESULTS_PER_CALL = 500  # 100 for sandbox, 500 for paid tiers
TO_DATE = datetime.strptime('2019-10-31 23:59','%Y-%m-%d %H:%M') # format YYYY-MM-DD HH:MM (hour and minutes optional)
FROM_DATE = datetime.strptime('2019-10-31 22:59','%Y-%m-%d %H:%M')  # format YYYY-MM-DD HH:MM (hour and minutes optional)
END_DATE = datetime.strptime('2019-10-15 00:00','%Y-%m-%d %H:%M')

MAX_RESULTS = 500  # Number of Tweets you want to collect

FILENAME = ''  # Where the Tweets should be saved
file_counter = 1

# Script prints an update to the CLI every time it collected another X Tweets
PRINT_AFTER_X = 100
```

```python


import yaml
config = dict(
    search_tweets_api=dict(
        account_type='premium',
        endpoint=f"https://api.twitter.com/1.1/tweets/search/{API_SCOPE}/{DEV_ENVIRONMENT_LABEL}.json",
        consumer_key=API_KEY,
        consumer_secret=API_SECRET_KEY
    )
)

with open('twitter_keys.yaml', 'w') as config_file:
    yaml.dump(config, config_file, default_flow_style=False)

    
import json
from searchtweets import load_credentials, gen_rule_payload, ResultStream

premium_search_args = load_credentials("twitter_keys.yaml",
                                       yaml_key="search_tweets_api",
                                       env_overwrite=False)

while TO_DATE>END_DATE:
    
   
    
    rule = gen_rule_payload(SEARCH_QUERY,
                            results_per_call=RESULTS_PER_CALL,
                            from_date=datetime.strftime(FROM_DATE,'%Y-%m-%d %H:%M'),
                            to_date=datetime.strftime(TO_DATE,'%Y-%m-%d %H:%M')
                            )
    
    
    
    
    rs = ResultStream(rule_payload=rule,
                      max_results=MAX_RESULTS,
                      **premium_search_args)
    
    TO_DATE = FROM_DATE
    
    ####Divide your time window using timedelta below
    
    FROM_DATE = (TO_DATE - timedelta(hours=1))
    
    
    
    FILENAME = "twitter_"+str(file_counter)+".jsonl"
    
    file_counter += 1


    with open(FILENAME, 'a', encoding='utf-8') as f:
        n = 0
        for tweet in rs.stream():
            n += 1
            if n % PRINT_AFTER_X == 0:
                print('{0}: {1}'.format(str(n), tweet['created_at']))
            json.dump(tweet, f)
            f.write('\n')
    
    time.sleep(3)
    
    
```
